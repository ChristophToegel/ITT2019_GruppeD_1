```python
import csv
reaction_time_distraction_attentive=[]
reaction_time_attentive=[]
reaction_time_pre_attentive=[]
reaction_time_distraction_pre_attentive=[]
with open('reaction_time_results.csv','r') as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter=';')
    for row in csv_reader:
        reaction_time = float(row["reaction_time"].replace(',','.'))
        stimuli = row["stimulus"]
        if stimuli == "AD":
            reaction_time_distraction_attentive.append(reaction_time)
        elif stimuli == "AN":
             reaction_time_attentive.append(reaction_time)
        elif stimuli=="PD":
            reaction_time_distraction_pre_attentive.append(reaction_time)
        elif stimuli=="PN":
            reaction_time_pre_attentive.append(reaction_time)
#print(reaction_time_distraction_attentive)
#print(reaction_time_attentive)
#print(reaction_time_pre_attentive)
#print(reaction_time_distraction_pre_attentive)
```

### T-test

```python
from scipy.stats import ttest_ind, ttest_rel, ttest_1samp

#Attentive vs pre-attentive
t_statistic, p_value = ttest_ind(reaction_time_attentive, reaction_time_pre_attentive)
print("p-value: %2.30f" %(p_value))
```

```python
#non-distracted vs distracted
t_statistic, p_value = ttest_ind(reaction_time_distraction_attentive, reaction_time_distraction_pre_attentive)
print("p-value: %2.30f" %(p_value))
```
